```python
# Inisiasi variabel awal

threshold = -1

X1 = 0.7
X2 = 0.8
X3 = 0.9

w14 = 0.5
w15 = 0.6

w24 = 0.3
w25 = 1.1

w34 = -1.0
w35 = 0.1

w46 = -1.1
w56 = -0.7

theta4 = 0.2
theta5 = 0.3
theta6 = 0.4

y6_true = 0

learning_rate = 0.1

```

```python
import numpy as np

def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

def sigmoid_derivative(x):
    return x * (1 - x)
```

## Melakukan Forward pass

```python
# Mencari nilai y4

num_y4 = (X1*w14 + X2*w24 + X3*w34 + (threshold*theta4))
y4 = sigmoid(num_y4)
y4
```

```python
# Mencari nilai y5

num_y5 = (X1*w15 + X2*w25 + X3*w35 + threshold*theta5)
y5 = sigmoid(num_y5)
y5
```

```python
# mencari nilai y6_hat

num_y6 = (y4*w46 + y5*w56 + threshold*theta6)
y6_hat = sigmoid(num_y6)
y6_hat
```

```python
# Nilai Error
err = y6_true - y6_hat
err
```

## Melakukan Backward pass / Backpropagation

```python
err_grad_y6_hat = sigmoid_derivative(y6_hat) * err
err_grad_y6_hat
```

```python
delta_w46 = learning_rate * err_grad_y6_hat * y4
delta_w46
```

```python
delta_theta4 = learning_rate * -1 * delta_y4
delta_theta4
```

```python
delta_y5 = sigmoid_derivative(y5) * err_grad_y6_hat * w56
delta_y5
```

```python
delta_theta5 = learning_rate * -1 * delta_y5
delta_theta5
```

```python
with open('modelMLP.pickle','rb') as f:
   modelMLP = pickle.load(f)
```

```python

with open('Vectorize.pickle','rb') as f:
   Vectorize = pickle.load(f)

with open('Transformer.pickle','rb') as f:
   Transformer = pickle.load(f)

# Untuk stemmer kata
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Melakukan preprocessing
def text_proses(teks):
  teks = teks.lower()
  teks = re.sub("@[A-Za-z0-9_]+"," ", teks) #Menghilangkan mention
  teks = re.sub("#[A-Za-z0-9_]+"," ", teks) #Menghilangkan hashtag
  teks = re.sub(r"\\[n|t]"," ",teks) #Menghilangkan \n
  teks = re.sub(r'\\x..', '', teks)
  teks = re.sub(r'&amp;', '', teks)
  teks = teks.strip()
  teks = re.sub(r"http\S+", " ", teks) #Menghilangkan link
  teks = re.sub(r"www.\S+", " ", teks) #Menghilangkan link
  teks = re.sub("[^A-Za-z\s']"," ", teks) #Menghilangkan yang bukan huruf
  teks = re.sub("rt"," ",teks) #Menghilangkan RT
  tokens = word_tokenize(teks)
  teks = ' '.join([word for word in tokens ])
  teks = stemmer.stem(teks)
  return teks
```
